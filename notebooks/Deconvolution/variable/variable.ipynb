```python
import numpy as np
from tnia.plotting.projections import show_xyz_max
```

```python
def ramp_3d(array, a, b):
    # Get the slice index range to apply the ramp to
    slice_range = range(min(a, b), max(a, b) + 1)

    # Calculate the ramp values
    ramp_values = np.linspace(0, 1, len(slice_range)).reshape(-1, 1, 1)

    # Set the ramp values in the slice range
    array[slice_range, :, :] = ramp_values

    return array
```

```python
arr = np.zeros((100,256,256))

test = ramp_3d(arr, 30, 70)

fig = show_xyz_max(test)
```
