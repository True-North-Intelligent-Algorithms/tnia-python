```python
import random
import albumentations as A
from tnia.plotting.plt_helper import imshow_multi2d
```

```python
from pathlib import Path
import json

tnia_images_path = Path(r'D:/images')
parent_path=Path(tnia_images_path / r'tnia-python-images/imagesc/2024_01_12_plant_roots')
train_path = parent_path / 'labels'

from tnia.deeplearning.dl_helper import collect_training_data

X, Y = collect_training_data(train_path, sub_sample=1, downsample=False)

print('type X ', type(X))
print('type Y ', type(Y))
```

```python
print(X[0].shape, Y[0].shape)
print(X[0].min(), X[0].max())
fig=imshow_multi2d([X[0], Y[0]], ['input', 'label'], 1,2)
```

```python
X_ = X[0].copy()
Y_ = Y[0].copy()
```

```python
aug = A.CenterCrop(p=1, height=155, width=191)

augmented = aug(image=X_, mask=Y_)

X_aug = augmented['image']
Y_aug = augmented['mask']

print(X_aug.shape, Y_aug.shape)

fig = imshow_multi2d([X[0], Y[0], X_aug, Y_aug], ['input', 'label','input aug', 'label aug'], 2,2, width=7, height=7)
```
