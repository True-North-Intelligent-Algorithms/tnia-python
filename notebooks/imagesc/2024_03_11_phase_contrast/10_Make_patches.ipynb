## Make patches

Make patches by calling the uber augmenter

```python
from skimage.io import imread
from pathlib import Path
from matplotlib import pyplot as plt
from tnia.plotting.plt_helper import imshow_multi2d
import napari
import numpy as np
from tnia.deeplearning.dl_helper import quantile_normalization

tnia_images_path = Path(r'D:/images')
parent_path=Path(tnia_images_path / r'tnia-python-images/imagesc/2024_03_11_phase_contrast')
label_path = Path(parent_path / r'labels')
image_path = Path(label_path / r'inputs')
mask_path = Path(label_path / r'ground_truth')
patch_path = Path(parent_path / r'patches')

image = imread(str(image_path / ("image3.tif")))
mask = imread(str(mask_path / ("image3.tif")))

fig = imshow_multi2d([image, mask], ["image", "mask"], 1, 2)
```

## Normalize image here

```python
image = quantile_normalization(image).astype(np.float32)
```

```python
import os
from tnia.deeplearning.augmentation import uber_augmenter

patch_path= parent_path / 'patches' 

if not os.path.exists(patch_path):
    os.mkdir(patch_path)
axes = 'YX'
sub_sample = 1

uber_augmenter(image, mask, patch_path, 'grains', 256, 40)

```
