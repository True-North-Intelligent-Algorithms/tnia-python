```python
from skimage.io import imread
from pathlib import Path
from matplotlib import pyplot as plt
from tnia.plotting.plt_helper import imshow_multi2d
import numpy as np
from tnia.deeplearning.dl_helper import random_label_cmap
from csbdeep.utils import normalize
from skimage.io import imread

tnia_images_path = Path(r'D:\images')
parent_path=Path(tnia_images_path / r'tnia-python-images/imagesc/2024_03_21_nuclei_segmentation')
label_path = parent_path / 'labels'
patch_path = parent_path / 'patches'

input_path = label_path / 'inputs0'
gt_path = label_path / 'ground truth0'
import os

# get all files in label_path
image_name = os.listdir(input_path)[0]
label_name = os.listdir(gt_path)[0]

print(image_name)
print(label_name)

# exit cell

# left upper quadrant index

image = imread(input_path / image_name)
ind = np.s_[:image.shape[0]//2,:image.shape[1]//2]
image = image[ind]
image = normalize(image,1,99.8, axis=(0,1))

labels = imread(gt_path / label_name)
labels = labels[ind]
fig = imshow_multi2d([image,labels],['input','predicted'],1,2, width=14, height=8,colormaps=['gray', random_label_cmap()])
```

```python
from tnia.deeplearning.augmentation import uber_augmenter
import os

if not os.path.exists(patch_path):
    os.mkdir(patch_path)
axes = 'YX'
sub_sample = 1

uber_augmenter(image, labels, patch_path, 'nuclei', 128, 500, do_random_sized_crop=True, do_random_gamma=True, do_elastic_transform=True)

```
