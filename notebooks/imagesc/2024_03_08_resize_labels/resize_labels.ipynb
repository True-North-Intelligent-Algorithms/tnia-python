## Resize labelling without interpolation artifacts

This notebooks shows how to resize a label while avoiding interpolation problems.  

The trick is to use nearest neighbor interpolation and set ```anti_anliasing``` to false. 

```
test2 = resize(test, (50,50), order=0, anti_aliasing=True)
```

````python
import raster_geometry as rg
````

````python
size = [100,100]
test = rg.circle(size, 20)
test = 5*test
````

````python
from matplotlib import pyplot as plt

plt.imshow(test)
````

````python
from skimage.transform import resize
import numpy as np
test2 = resize(test, (50,50), order=0, anti_aliasing=True)
plt.imshow(test2)
np.unique(test2)
````

````python
test2 = resize(test, (50,50), order=0, anti_aliasing=False)
plt.imshow(test2)
np.unique(test2)
````
