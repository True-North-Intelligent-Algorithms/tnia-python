```python
from pathlib import Path
from tnia.plotting.projections import show_xyz_max
from skimage.io import imread

parent_path = r"D:\images\tnia-python-images\imagesc\2024_09_09_clij2_rl_nc"
img_name = "img.tiff"
psf_name = "psf.tiff"

img_path = Path(parent_path) / img_name
psf_path = Path(parent_path) / psf_name

img = imread(img_path)
psf = imread(psf_path)

fig = show_xyz_max(img, sxy = 1, sz = 5)
```

```python
from clij2fft.richardson_lucy import richardson_lucy_nc, richardson_lucy

decon_nc = richardson_lucy_nc(img, psf, 100)
decon = richardson_lucy(img, psf, 100)
```

```python
fig = show_xyz_max(decon, sxy = 1, sz = 5)
fig = show_xyz_max(decon_nc, sxy = 1, sz = 5)
```

```python

from tnia.plotting.projections import show_xy_zy_slice
import numpy as np
# crop to visualize a little better

s = np.s_[:, :240, :240]

fig = show_xy_zy_slice(img[s], 80, 80, 11, sxy = 1, sz = 5, figsize=(7,5))
fig = show_xy_zy_slice(decon[s], 80, 80, 11, sxy = 1, sz = 5, figsize=(7,5))
fig = show_xy_zy_slice(decon_nc[s], 80, 80, 11, sxy = 1, sz = 5, figsize=(7,5))


```
