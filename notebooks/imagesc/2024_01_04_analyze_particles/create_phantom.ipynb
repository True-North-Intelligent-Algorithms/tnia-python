```python
import raster_geometry as rg
from skimage.io import imsave 
from tnia.simulation.phantoms import add_small_to_large_2d
import numpy as np
from tnia.plotting.plt_helper import imshow2d
```

```python
xdim = 1024
ydim = 512
img = np.zeros((ydim, xdim))

r = 81
# add a small circle to the image
for xc in range(100, xdim-50, 200):
    yc = 100
    circle = rg.circle((r*2, r*2), r).astype('int8')*255
    add_small_to_large_2d(img, circle, xc, yc)
    r=r-15

r = 81
for xc in range(100, xdim-50, 200):
    yc = 350
    rectangle=rg.rectangle((r*2, r*2), (r,10)).astype('int8')*255
    #square = rg.square((r*2, r*2), r*2).astype('int8')*255
    add_small_to_large_2d(img, rectangle, xc, yc)
    r=r-15
fig = imshow2d(img)

```

```python
imsave('./phantom.tiff', img)
```
