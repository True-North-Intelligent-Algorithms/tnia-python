## Citrus Lesion LAB Color Space


## Read the data

And crop out the scale bar

```python
from skimage.io import imread
from tnia.plotting.plt_helper import imshow2d, imshow_multi2d
from skimage.filters import gaussian
from skimage.color import rgb2lab
from skimage.filters import threshold_otsu, threshold_mean, threshold_triangle, threshold_yen
from tnia.segmentation.separate import separate_touching
from skimage.morphology import remove_small_objects
import numpy as np

from pathlib import Path

im_name = "20240108_123559"

tnia_images_path = Path("D:/images/tnia-python-images")
parent_path = tnia_images_path / "imagesc/2024_02_29_citrus_lesion"
im = imread(str(parent_path / (im_name+".jpg")))
           
fig = imshow2d(im)
```

## Convert to LAB color space

And show the converted channels and histograms



```python
# convert to HSB
im_lab = rgb2lab(im)

fig = imshow_multi2d([im_lab[:,:,0], im_lab[:,:,1], im_lab[:,:,2]], ['Lightness', 'Red to Green', 'Blue to Yellow'], 1, 3, width=15)

hists = []
for i in range(3):
    hists.append(np.histogram(im_lab[:,:,i], bins=100))

fig = imshow_multi2d([im_lab[:,:,0], im_lab[:,:,1], im_lab[:,:,2]], ['Lightness', 'Red to Green', 'Blue to Yellow'], 1, 3, width=15, height=5, plottypes=3*['hist'])

```

## Threshold to create first guess for labels

```python
L, a, b = im_lab[:,:,0], im_lab[:,:,1], im_lab[:,:,2]

binary_L = L > threshold_otsu(L)
binary_a = a > threshold_otsu(a)
binary_b = b > threshold_otsu(b)

fig = imshow_multi2d([binary_L, binary_a, binary_b], ['Lightness', 'Red to Green', 'Blue to Yellow'], 1, 3, width=15)
```

```python
binary_and = binary_a & binary_b

fig = imshow2d(binary_and)
```
