```python
from skimage.io import imread
from pathlib import Path
from matplotlib import pyplot as plt
import napari
import numpy as np

tnia_images_path = Path("D:/images/tnia-python-images")
parent_path = tnia_images_path / "imagesc/2024_03_05_sem_grain_size"

image = imread(str(parent_path / ("SEM.tif")))
segmented = imread(str(parent_path / ("segmented.tif")))

outlines = np.zeros_like(segmented)
outlines[segmented==1]=1

label = outlines[:512,:512]

plt.imshow(label)
```

```python
viewer = napari.Viewer()
viewer.add_image(image[:512,:512], name='image')
viewer.add_labels(label, name='labels')
```

```python
from skimage.io import imsave
imsave(parent_path / "label.tif", label.astype('uint16'))
```

```python
imsave(parent_path / "image.tif", image[:512,:512].astype('uint16'))
```
