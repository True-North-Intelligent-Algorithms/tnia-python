```python
from czifile import CziFile
import tifffile
import os
import napari
```

```python
parent_path = r'D:\images\tnia-python-images\imagesc\2025_09_08_File_Format_Examples'

def what_am_I(file_path):
    if file_path.endswith('.czi'):
        try:
            czi_file = CziFile(file_path)
            return 'czi', czi_file.shape, czi_file.axes
        except Exception as e:
            return 'czi_error, maybe czi reader is not installed', str(e), None
    elif file_path.endswith('.tif') or file_path.endswith('.tiff'):
        img = tifffile.imread(file_path)
        return 'tiff', img.shape, None
    else:
        return 'idontknow', None, None

# get all files in the directory
files = os.listdir(parent_path)

for f in files:
    print(f)
    print(what_am_I(os.path.join(parent_path, f)))
```

```python
def load_czi_image(file_path):
    with CziFile(file_path) as czi:
        img = czi.asarray()
        axes = czi.axes
    return img, axes
```

```python
parent_path = r'D:\images\tnia-python-images\imagesc\2025_09_08_Alex_File_Format_Examples'
im_name = r'2D_1C_1201T_epifluo_LSM980.czi'

parent_path = r'D:\images\tnia-python-images\imagesc\2025_05_20_3D_Segmentation'
im_name = r'Image 6_Subset-pos02_t1-35.czi'

img, axes = load_czi_image(os.path.join(parent_path, im_name))
print(f'Image shape: {img.shape}, Axes: {axes}')
```

```python
im_name2 = r'3D_4C_Airyscan_LSM980.czi'
im_name2 = r'Image 6_Subset-pos02_t36-72.czi'
img2, axes = load_czi_image(os.path.join(parent_path, im_name2))
print(f'Image shape: {img2.shape}, Axes: {axes}')
```

```python
print(type(img))
print(type(axes))

# Remove any axis where shape=1
new_shape = []
new_axes = []
for dim, ax in zip(img.shape, axes):
    if dim != 1:
        new_shape.append(dim)
        new_axes.append(ax)

print(f'New shape: {new_shape}, New Axes: {new_axes}')
```
