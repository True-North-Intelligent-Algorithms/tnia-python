```python
import os
import numpy as np
from pathlib import Path
from skimage.io import imread, imsave
```

```python

parent_path = Path(r'/home/bnorthan/images/tnia-python-images/imagesc/2024_10_14_cellpose_masks')
mask_path = Path(r'/home/bnorthan/images/tnia-python-images/imagesc/2024_10_14_cellpose_masks/images_and_masks')
#/home/bnorthan/images/tnia-python-images/imagesc/2024_10_14_cellpose_masks/images_and_masks/
annotation_path = parent_path / 'annotations' / 'class0'

if not os.path.exists(annotation_path):
    os.makedirs(annotation_path)
```

```python
# get all npy files
mask_files = list(mask_path.glob('*.npy'))

len(mask_files)
```

```python
for mask_file in mask_files:
    temp = np.load(mask_file, allow_pickle=True)
    mask = temp.item().get('masks')

    tif_file = Path(str(mask_file).replace('_seg.npy', '.tif'))
    base_name = os.path.basename(tif_file)
    tif_file = annotation_path / base_name
    print(tif_file)
    imsave(tif_file, mask)
    print(mask.shape)
```
