```python
from skimage.io import imread
import czifile
from pathlib import Path

parent_path = r'D:\images\tnia-python-images\imagesc\2025_05_20_3D_Segmentation'
image_name = 'Image 6_Subset-pos02_t36-72.czi'

image = czifile.imread(Path(parent_path) / image_name)

print(type(image), print(image.shape), print(image.dtype))

```

```python
import numpy as np

image = np.squeeze(image)
print(type(image), print(image.shape), print(image.dtype))
```

```python
import napari

viewer = napari.Viewer()
viewer.add_image(image, name=image_name, scale = [1, 1, 10, 1, 1])
```

```python
image_crop = image[:,:,:,500:600, 500:600]
print(image_crop.shape)
viewer = napari.Viewer()
viewer.add_image(image_crop, name='image_crop', scale = [1, 1, 1, 1, 1])
```

```python
from skimage.io import imsave

# loop through the channels
for i in range(image.shape[0]):
    # loop through the time points
    for j in range(image.shape[1]):
        stack = image[i, j, :, :, :]
        print(stack.shape)
        name = Path(parent_path) / f'{image_name}_c{i}_t{j}.tif'
        imsave(name, stack)
```
