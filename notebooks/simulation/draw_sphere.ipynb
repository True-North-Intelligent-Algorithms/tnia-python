## Test sphere and ellipse rendering

This notebook can be used to test how spheres are rendered into the pixelated image space by the raster_geometry tooklkit. 

Below 3 spheres are drawn with slightly different sizes.  Note that the differences in the resulting pixelated spheres are subtle.  Spheres of slightly larger radii have the same 'diameter' as a slightly smaller one, but larger 'volume' (larger pixel count).  

```python

import numpy as np
import raster_geometry as rg

sphere_a = rg.sphere(12, 5).astype(np.float32)
sphere_b = rg.sphere(12, 5.5).astype(np.float32)
sphere_c = rg.sphere(12, 5.55).astype(np.float32)

'''
import napari
viewer = napari.Viewer()
viewer.add_image(sphere_a, name='sphere_a')
viewer.add_image(sphere_b, name='sphere_b')
viewer.add_image(sphere_c, name='sphere_c')
viewer.show()
'''
from tnia.plotting.projections import show_xy_zy_max
fig = show_xy_zy_max(sphere_a)
fig = show_xy_zy_max(sphere_b)
fig = show_xy_zy_max(sphere_c)
```
