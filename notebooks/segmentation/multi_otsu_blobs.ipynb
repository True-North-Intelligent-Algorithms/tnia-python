```python
from skimage.io import imread
from skimage.filters import threshold_otsu
import matplotlib.pyplot as plt
import numpy as np
```

```python
im_name=r"D:\images\tnia-python-images\segmentation\ijsamples\blobs.tif"
```

```python
im=imread(im_name)
plt.imshow(im)
```

```python
threshold_0=threshold_otsu(im)
binary_0=im<=threshold_0
binary_1=im>threshold_0

threshold_00=threshold_otsu(im[binary_0])
threshold_10=threshold_otsu(im[binary_1])

multi=np.zeros(im.shape)
multi[im<threshold_00]=0
multi[(threshold_00<im) & (im<threshold_0)]=1
multi[(threshold_0<im) & (im<threshold_10)]=2
multi[im>threshold_10]=3

#threshold_
#binary2=im[binary1]>threshold_otsu(im[binary1])
fig=plt.imshow(multi)
```
