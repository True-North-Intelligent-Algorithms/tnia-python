```python
from tnia.io.io_helper import open_zero_padded_numbered_stack
from tnia.plotting.projections import show_xy_zy_max, show_xy_zy_slice
from tnia.plotting.plt_helper import create_rgb, color_dictionary 
```

```python
### Use skimage cells data

from skimage.data import cells3d

image_3D = cells3d()
print(f'image_3D.shape: {image_3D.shape}')
```

```python
rgb = create_rgb(image_3D)

print(image_3D.min(), image_3D.max())
print(rgb.min(), rgb.max())
```

```python
fig = show_xy_zy_max(rgb, sz=3, figsize=(7,4))
fig = show_xy_zy_slice(rgb, 100, 100,30,sz=3, figsize=(7,4), show_cross_hairs=False)
fig = show_xy_zy_slice(rgb, 100, 100,30,sz=3, figsize=(7,4), show_cross_hairs=True, vmax=.3)
```

```python
waves = ['FITC', 'DAPI']

print(color_dictionary)

rgb2 = create_rgb(image_3D, waves=waves)

fig = show_xy_zy_slice(rgb2, 100, 100,30,sz=3, figsize=(7,4), show_cross_hairs=True, vmax=.3)
```

```python
waves = ['purple', 'cyan']

rgb3 = create_rgb(image_3D, waves=waves)

fig = show_xy_zy_slice(rgb3, 100, 100,30,sz=3, figsize=(7,4), show_cross_hairs=True)

rgb4 = create_rgb(image_3D, waves=waves, vmin=[0,0], vmax=[.3, .4], gamma=[1,1])

fig = show_xy_zy_slice(rgb4, 100, 100,30,sz=3, figsize=(7,4), show_cross_hairs=True)
```
