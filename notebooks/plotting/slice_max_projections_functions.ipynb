```python
### Use skimage cells data

from skimage.data import cells3d

image_3D = cells3d()
print(f'image_3D.shape: {image_3D.shape}')

nuclei_3D = image_3D[:,1,...]  # pick channel, make sure ZYX dim order
print(f'nuclei_3D.shape: {nuclei_3D.shape}')
```

```python
from tnia.plotting.projections import show_xyz_max, show_xyz_max_slabs

fig = show_xyz_max(nuclei_3D, figsize=(5,5))
```

```python
fig = show_xyz_max_slabs(nuclei_3D, x = [0,50], y = [0,50], z = [0,60], figsize=(5,5))
```
