## Multi-channel projections

Link to the images for this example is [here](https://www.dropbox.com/scl/fo/bvm753vpeqegfvw8ywuv9/h?rlkey=qnr2f75dk0oujxvnz2o35pogt&dl=0)

First we load the data and tranform the three channels to an RGB image

```python
from skimage.io import imread
from tnia.plotting.projections import show_xyz_max, show_xy_xz_slice, show_xy_zy_slice
from tnia.plotting.interactive_projections import show_xyz_slice_interactive
import numpy as np
import os
from skimage import color

path_=r'/home/bnorthan/images/tnia-python-images/deconvolution/c.elegans/'

# file names
names = ['CElegans-CY3.tif', 'CElegans-DAPI.tif', 'CElegans-FITC.tif']
# load images
ims = [imread(path_+'/'+name) for name in names]

# make a color map (red, blue, green)
maps = [[1,0,0], [0,0,1], [0,1,0]]

# make a color image
im = np.zeros((ims[0].shape[0], ims[0].shape[1], ims[0].shape[2], 3), dtype=np.float32)

# loop over images and maps and add to color image
for im_, map in zip(ims, maps):
    temp = color.gray2rgb(im_) * map
    temp = temp/temp.max()
    im+=temp

# show the image
fig = show_xyz_max(im, sxy=0.1, sz=0.3, figsize=(10,10))
```

# XY_ZY slice projection

There are multiple types of projections we can do.  Here do XY_ZY slice projection

```python
fig = show_xy_zy_slice(im,  300, 300, 50, sxy=.1, sz=.3,figsize=(15,10))
fig.suptitle('CElegans', fontsize=15)
```

# And an interactive slice projection
