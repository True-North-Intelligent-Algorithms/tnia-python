## Static projections

There are some great tools for dynamic projections like Robert Haase's [StackView](https://pypi.org/project/stackview/) and the ```tnia-python``` library also has some dynamic projection tools (see ```20_interactive_stacks```)

However static projections are also very useful.  One advantage of static projections is that they render on the github and the web (when dynamic widgets sometimes do not).  Another advantage is that they are useful for persistent figures.  



## Load Test Image

We will use the Nuclei channel of the ```skimage.data.cells3d``` sample.  

````python
### Use skimage cells data

from skimage.data import cells3d

image_3D = cells3d()
print(f'image_3D.shape: {image_3D.shape}')

nuclei_3D = image_3D[:,1,...]  # pick channel, make sure ZYX dim order
print(f'nuclei_3D.shape: {nuclei_3D.shape}')
````

## Center Slice Projection

Center slice projection is a good way to start exploring the data.  We can pass ```sxy``` (scale xy) and ```sz``` (scale z) to make sure the scale is correct. 

````python
from tnia.plotting.projections import show_xyz_slice_center

fig = show_xyz_slice_center(nuclei_3D, figsize = (7,7), sxy=.1, sz=.3)
fig.suptitle('Nuclei 3D center slice projections', fontsize=16)
````

## Slice Projection

````python
from tnia.plotting.projections import show_xyz_slice

fig = show_xyz_slice(nuclei_3D, 50, 60, 34, figsize = (7,7), sxy=.1, sz=.3)
fig.suptitle('Nuclei 3D slice projections', fontsize=16)
````

## Max Projection

Max projections can be useful to see everything in the image at once. They are especially useful for images with sparse structure, however can be confusing for images with dense structure (in the figure below the zy and xz projections are quite busy).

````python
from tnia.plotting.projections import show_xyz_max

fig = show_xyz_max(nuclei_3D, figsize = (7,7), sxy=.1, sz=.3)
````

## Show only xy_zy or xy_xz projections

Sometimes it's useful to show only the xy and zy or xy and xz projection.  For example if doing comparisons between different images.  

````python
from tnia.plotting.projections import show_xy_zy_slice, show_xy_xz_slice, show_xy_zy_max

fig = show_xy_zy_slice(nuclei_3D, 50, 60, 30, figsize = (7,4.5), sxy=.1, sz=.3)
fig.suptitle('Nuclei 3D XY-ZY slice projections', fontsize=16)
fig = show_xy_zy_max(nuclei_3D, figsize = (7,4.5), sxy=.1, sz=.3)
fig.suptitle('Nuclei 3D XY-ZY max projections', fontsize=16)
````

## Changing colormap

````python
fig = show_xy_zy_slice(nuclei_3D, 50, 60, 30, figsize = (7,7), sxy=.1, sz=.3, colormap = 'gray')
fig = show_xy_zy_slice(nuclei_3D, 50, 60, 30, figsize = (7,7), sxy=.1, sz=.3, colormap = 'inferno')
fig = show_xy_zy_slice(nuclei_3D, 50, 60, 30, figsize = (7,7), sxy=.1, sz=.3, colormap = 'Spectral')
````

## Saving figures

Simply use ```fig.savefig(...)```

````python
fig = show_xy_zy_slice(nuclei_3D, 50, 60, 30, figsize = (7,4.5), sxy=.1, sz=.3)

fig.savefig('nuclei_3D.png')
````
