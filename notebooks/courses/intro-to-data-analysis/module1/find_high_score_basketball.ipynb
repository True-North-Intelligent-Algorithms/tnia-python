## Problem

Load the Jalen_Bunson.csv file and try and use the Pandas library and/or other Python techniques to find Jalen Brunson's high score

## Method 1

```python
import pandas as pd
```

```python
my_data=pd.read_csv('Jalen_Brunson.csv')
```

```python
def make_str_number(x):
    try:
        y=int(x)
    except:
        y=0
    return y
```

```python
my_data['PTS2']=my_data['PTS'].apply(make_str_number)
```

```python
my_data[my_data.PTS2==my_data.PTS2.max()]
```
