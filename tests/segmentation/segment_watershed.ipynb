```python
from skimage.io import imread
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu
from tnia.segmentation.separate import separate_touching
from tnia.segmentation.rendering import draw_label_outline
%matplotlib inline
```

```python
blobs = imread('../../images/blobs.tif')
```

```python
thresh = threshold_otsu(blobs)
binary = blobs > thresh
plt.imshow(binary)
```

```python
separated = separate_touching(binary, 10, 1)
plt.imshow(separated)
```

```python
outlined = draw_label_outline(blobs, separated)
plt.imshow(outlined)
```
